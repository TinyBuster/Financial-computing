```python
import requests
from html.parser import HTMLParser

class MyHTMLParser(HTMLParser):
    def handle_starttag(self, tag, attrs):
        global innew, intable, inrow, incell, rowlist
        if tag=="th":
            innew=1
        if tag=="table":
            intable=1
        if tag=="tr": # start of a new row
            inrow=1
            rowlist=[]
        if tag=="td":
            incell=1
    def handle_endtag(self, tag):
        global innew, current_future, intable, inrow, incell, rowlist, L
        if tag=="th":
            innew=0
        if tag=="table":
            intable=0
        if tag=="tr": 
            inrow=0
            if len(rowlist) == 10: # record data to future_price_data
                data_key = (current_future, rowlist[0], rowlist[1])
                previous_day = (rowlist[7], rowlist[8], rowlist[9])
                data_value = (rowlist[2], rowlist[3], rowlist[4], rowlist[5], rowlist[6], previous_day)
                future_price_data[data_key] = data_value
        if tag=="td":
            incell=0
    def handle_data(self, data):
        global current_future, intable, inrow, incell, rowlist
        if innew:
            current_future = data
        if intable and inrow and incell:
            rowlist.append(data) # add data to row list
            
                 
innew=0
intable=0
inrow=0
incell=0
rowlist=[]

future_price_data = {} # my dictionary
parser = MyHTMLParser()

for i in range(1,10):
    url = "http://www.mrci.com/ohlc/2008/08020" + str(i) + ".php"
    web = requests.get(url = url)
    parser.feed(web.text)  

for i in range(10,30):
    url = "http://www.mrci.com/ohlc/2008/0802" + str(i) + ".php"
    web = requests.get(url = url)
    parser.feed(web.text) 

future_price_data # print my dictionary
```
